## 예제 17. 인플루언서의 허영심 (Vanity)을 측정하기

- 각 인플루언서의 허영심 점수 ( = 링크 인 바이오 수 * 5 + 업로드 된 셀카 수)를 측정하여 정렬해봅시다.
    - 링크 인 바이오는 흔히 볼 수 있는 인플루언서의 프로필에서 외부로 연결된 링크입니다. 예를 들어 인스타그램 내 A 연예인의 프로필에 입력되어 있는 그 연예인의 유튜브 채널은 링크 인 바이오죠.

```python
class Influencer:
    def __init__(self, num_selfies, num_bio_links):
        self.num_selfies = num_selfies
        self.num_bio_links = num_bio_links

    def __repr__(self):
        return f"({self.num_selfies}, {self.num_bio_links})"


# dont touch above this line

# influencer의 vanity를 계산하여 반환하는 함수
def vanity(influencer): 
    return influencer.num_selfies + influencer.num_bio_links * 5

# influencer의 list로 입력된 데이터를
# vanity 함수의 결과값을 기준값으로 하여 정렬합니다.
def vanity_sort(influencers): 
    return sorted(influencers, key = vanity)

# 예시로, 3명의 influencer가 있다고 가정해봅시다.
influencers = [Influencer(10, 2), Influencer(5, 5), Influencer(8, 1)]
vanity_sort(influencers)
```

## Bubble Sort (버블 정렬)

근데, `sorted()` 처럼 이미 구현된 알고리즘이 있다면, 왜 굳이 알고리즘 구현을 처음부터 해봐야 하는걸까요? 그건 알고리즘 구현이 작동 방식의 이해에 아주 좋기 때문입니다. 그리고 아주 흥미롭기 때문이죠. (저한테는요.)

버블 정렬은 목록 내에 인접한 두 자료를 비교하고 필요 시 위치를 바꾸는 방식의 정렬 방법입니다. (잘 모르겠지만…) 정렬 과정에서 원소들의 움직임이 수면으로 올라오는 기포의 모양과 비슷하다고 붙여진 이름입니다. 기본적인 알고리즘의 흐름은 다음과 같습니다.

1. 최초에 반복문 진입을 위해 `정렬함` 변수를 `참`으로 설정합니다.
2. `종료 지점`은 `주어진 목록의 길이`와 동일하게 설정합니다.
3. `정렬함`이 `참` 일 때 아래 반복문을 수행합니다.
    1. 우선 `정렬함`을 다시 `거짓`으로 바꿉니다.
    2. 첫 번째 자료와 두 번째 자료를 비교하여 필요 시 위치를 바꿉니다. 
    3. 두 번째 자료와 세 번째 자료를 비교하여 필요 시 위치를 바꿉니다.
    4. … 위 행위를 `종료 지점`까지 반복합니다. 만약 위치가 바뀐 자료가 있다면, 아직 자료가 완벽하게 정렬되지 않았기 때문에 `정렬함`을 `참`으로 바꿔 반복문이 다시 수행되도록 합니다.
    5. `종료 지점`을 1 뺍니다. (매 정렬 시 정렬의 마지막 데이터는 항상 가장 큰 숫자가 배치됩니다.)
4. 정렬된 목록을 반환합니다.

## 예제 18. 버블 정렬 구현

- 이제 알고리즘의 흐름을 알았으니 구현해봅시다.

```python
def bubble_sort(nums):
    swapped = True
    end = len(nums)
    while swapped == True:
        swapped = False
        for i in range(1, end):
            if nums[i - 1] <= nums[i]:
                continue
            nums[i - 1], nums[i] = nums[i], nums[i - 1]
            swapped = True
            print(nums)
        end -= 1
    return nums                

# 예시로, 무작위로 배치된 1부터 10까지의 숫자 리스트가 있다고 가정해봅시다.
random_nums = [3, 1, 4, 5, 2, 10, 9, 6, 8, 7]
bubble_sort(random_nums)
```

- 이처럼 버블 정렬은 매 회차 N개의 자료를 비교 및 교환하는 작업을 자료의 개수(N)만큼 반복하기 때문에 시간 복잡도는 O(n^2) 입니다. 반면 아무리 최적의 자료 (이미 정렬된 자료)가 들어와도 N개의 자료를 비교 및 교환하는 작업을 수행해야 하기 때문에 $\Omega(n)$ 입니다.
- 그럼 이걸 왜 쓰냐고요? 안 씁니다. 단지 버블 정렬은 단순하기 때문에 정렬이 이런 방식으로 일어난다는 것을 이해하는데 아주 좋은 알고리즘입니다.

## Merge Sort (합병 정렬)

합병 정렬은 일종의 재귀적인 정렬 알고리즘입니다. 버블 정렬보다 아주 약간 빠르죠. 기본적으로는 입력된 자료를 잘게 쪼개서 정렬 후 합치고 다시 정렬 후 합치고 하는 방식으로 정렬합니다 (재귀적 분할 정복 알고리즘 (Recursive Divide and Conquer Algorithm)). 언급한 것과 같이 합병 정렬은 `자료를 분할하면서 다시 분할된 자료로 정렬을 재귀적으로 호출하는 함수`와 `두 자료를 정렬하면서 합병하는 함수`로 이뤄져 있습니다.

의사 코드를 한 번 보죠.

1. 분할 및 재귀 호출 함수
    1. 입력된 자료의 길이가 2보다 짧으면 이미 정렬되었거나, 빈 자료입니다. 그대로 반환합니다.
    2. 반면, 길이가 2 이상이면 자료를 반으로 잘라, `왼쪽`과 `오른쪽`으로 구분한 뒤, 왼쪽에 대해서 다시 자신을 호출하여 `정렬된 왼쪽` 자료를 만들고, 오른쪽도 동일하게 수행해 `정렬된 오른쪽` 자료를 만듭니다.
    3. 마지막으로 `정렬된 왼쪽` 자료와 `정렬된 오른쪽` 자료를 이용해 `합병 합수`를 실행하고 결과를 반환합니다.
2. 정렬 합병 함수
    1. 결과를 저장하기 위한 빈 자료를 만듭니다.
    2. 인덱싱을 위한 변수를 선언하고, 두 자료를 아래 쪽부터 비교하며 작은 숫자를 계속 결과 자료에 차곡차곡 쌓는 식으로 정리합니다.
    3. 만약 한 쪽 자료가 모두 사용되었다면, 남은 자료는 그대로 결과 자료에 쌓습니다.
    4. 정리된 자료를 반환합니다.

## 예제 19. 합병 정렬 구현

- 사실 무슨 말인지 잘 모르겠습니다. 코드를 보죠.

```python
def merge_sort(nums):
    # 1-a 항
    if len(nums) < 2:
        return nums

    # 1-b 항
    left = nums[:len(nums)//2]
    right = nums[len(nums)//2:]
    sorted_left_side = merge_sort(left)
    sorted_right_side = merge_sort(right)

    # 1-c 항
    return merge(sorted_left_side, sorted_right_side)

def merge(first, second):
    # 2-a 항
    final = []

    # 2-b 항
    i = j = 0
    while i < len(first) and j < len(second):
        if first[i] <= second[j]:
            final.append(first[i])
            i += 1
        else:
            final.append(second[j])
            j += 1

    # 2-c 항
    while i < len(first):
        final.append(first[i])
        i += 1
    while j < len(second):
        final.append(second[j])
        j += 1

    # 2-d 항
    print(f"first: {first}, second: {second}, final: {final}")
    return final

# 예시로, 무작위로 배치된 1부터 10까지의 숫자 리스트가 있다고 가정해봅시다.
random_nums = [3, 1, 4, 5, 2, 10, 9, 6, 8, 7]
merge_sort(random_nums)
```

- 합병 정렬의 시간 복잡도는 $O(N logN)$ 입니다. 버블 정렬은 $O(n^2)$ 이기 때문에, 합병 정렬이 버블 정렬보다 빠릅니다. 다만, 재귀적 호출을 메인으로 하는 합병 정렬 알고리즘은 메모리 (RAM)을 상당히 많이 사용한다는 단점이 있습니다.

## Insertion Sort (삽입 정렬)

삽입 정렬은 한 번에 정리된 자료를 만들어냅니다. 알고리즘 구조 상 O(n^2)의 속도를 가지기 때문에 큰 규모의 데이터에서는 합병 정렬보다는 비효율적이지만, 적은 규모의 데이터에서는 합병 정렬보다 빠르게 동작합니다. 그리고 메모리도 적게 먹죠. 의사 코드를 한 번 볼까요.

1. 데이터의 두 번째 값부터 시작합니다. 매 인덱스마다 아래 반복문을 수행합니다.
    1. j에 현재 인덱스를 저장합니다.
    2. j가 0보다 크고, j-1 번째 항목이 j 번째 항목보다 값이 클 때 아래 반복문을 수행합니다.
        1. j-1 번째 항목과 j 번째 항목을 서로 바꿉니다.
        2. j를 1만큼 줄입니다.
2. 정렬이 완료된 데이터를 반환합니다.

```python
def insertion_sort(nums):
    if len(nums) < 2: return nums
    for j in range(1, len(nums)):
        while j > 0 and nums[j - 1] > nums[j]:
            nums[j - 1], nums[j] = nums[j], nums[j - 1]
            j -= 1
        print(nums)
    return nums

# 예시로, 무작위로 배치된 1부터 10까지의 숫자 리스트가 있다고 가정해봅시다.
random_nums = [3, 1, 4, 5, 2, 10, 9, 6, 8, 7]
insertion_sort(random_nums)
```

- 삽입 정렬은 적은 데이터가 입력되었을 때 합병 정렬(Merge Sort)과 퀵 정렬(Quick Sort)보다 빠르게 동작합니다. 또한 일부만 정렬된 데이터들에서도 유용하게 동작하죠. 그래서 보통 10개 내외의 자료가 주어졌을 때는 삽입 정렬을 사용하고, 그 이상의 데이터에 대해서는 퀵 정렬을 사용하는 등의 방식으로 구현합니다.

## Quick Sort (퀵 정렬)

퀵 정렬은 실제 정렬 구현에서 널리 쓰이는 효율적인 알고리즘입니다. 합병 정렬과 비슷하게 자료를 나눠서 정리 및 다시 합치는 재귀적 분할 정복 알고리즘 (Recursive Divide and Conquer Algorithm)을 사용합니다. 

의사 코드를 봅시다.

1. 우선 주어진 자료에서 순서 배치의 기준이 될 기준점 (Pivot)을 설정합니다. 보통 맨 마지막 원소를 지정합니다.
2. 자료를 계속 비교하면서, 기준점보다 작은 숫자는 앞으로, 기준점보다 큰 숫자는 뒤로 보내면서 자리를 바꿉니다.
3. 1회 정리가 끝나면, 피벗을 두 그룹의 중앙으로 이동시키고, 양쪽 그룹에 대해서 동일한 과정을 반복합니다.

## 예제 20. 퀵 정렬 구현

```python
def quick_sort(nums, low=0, high=None):
    # 인덱스에 대한 입력이 없으면, low를 0, high를 가장 마지막 인덱스로 설정
    if high is None:
        high = len(nums) - 1

    # low가 high보다 낮을 때만 정렬이 끝나지 않은 시점이므로 반복문 실행
    if low < high:
        # partition 함수를 통해 자료를 정리하고 중간 지점 인덱스 정보를 받아옴
        middle = partition(nums, low, high)

        # low ~ middle - 1 (왼쪽/앞쪽) 자료를 다시 정렬
        quick_sort(nums, low, middle - 1)
        
        # middle ~ high (오른쪽/뒤쪽) 자료를 다시 정렬
        quick_sort(nums, middle, high)

        print(nums)
    return nums

def partition(nums, low, high):
    # pivot을 자료 내 가장 마지막 데이터로 설정
    pivot = nums[high]
    i = low - 1
    for j in range(low, high):
        if nums[j] < pivot:
            i += 1
            nums[j], nums[i] = nums[i], nums[j]
    nums[i + 1], nums[high] = nums[high], nums[i + 1]
    return i + 1

# 예시로, 무작위로 배치된 1부터 10까지의 숫자 리스트가 있다고 가정해봅시다.
random_nums = [3, 1, 4, 5, 2, 10, 9, 6, 8, 7]
quick_sort(random_nums)
```

퀵 정렬은 평균적인 경우에 $\Theta(n\log(n))$의 복잡도를 가지지만 최악의 경우 $O(n^2)$까지 효율이 낮아집니다. 이는 Partition 함수가 얼마나 반복되냐의 차이입니다.

## 퀵 정렬 개선하기

최악의 상황에서 퀵 정렬의 낮은 효율을 개선하기 위해 보통 2가지 방식을 적용합니다.

- 정렬 진행 전 모든 데이터를 무작위로 한 번 섞은 다음 정렬을 시행합니다. $O(n)$
- 중앙값을 능동적으로 찾아 해당 회차 정렬의 Pivot으로 설정합니다. $O(1)$

물론 최초 무작위 정렬 시 만약 데이터가 완전 역순으로 정렬되면 효율성이 낮아지는 것이 아닌지에 대한 의문이 있을 수 있으나, 무작위로 섞는 과정에서 완전 역순 정렬의 확률은 데이터의 크기에 따라 계승적으로 감소합니다. $\frac{1}{3!} = 1/6 \rightarrow \frac{1}{4!} = 1/12  \rightarrow \frac{1}{5!} = 1/60  \rightarrow ...$ 거의 불가능하죠.

그리고 데이터의 첫번째, 중간 위치, 마지막 데이터를 선정하여 그 세 값 중 중앙값을 피벗으로 설정합니다. 무작위성이 불필요하게 되죠.

## Selection Sort (선택 정렬)

선택 정렬은 버블 정렬과 비슷하게 반복적으로 두 요소를 선택한 후 자리를 바꾸는 방식입니다. 다만, 한 회차에 한 번의 자리 변경만 진행하기 때문에 조금 더 효율적입니다. 의사 코드를 한 번 볼까요.

- 자료의 매 인덱스마다 다음 반복문을 수행합니다.
    - 현재 인덱스를 가장 작은 숫자 인덱스로 설정합니다.
    - 현재 인덱스 이후부터 마지막까지 자료를 순회하면서, 어떤 숫자가 현재 인덱스의 숫자보다 작다면, 그 숫자를 가장 작은 숫자 인덱스로 설정
    - 순회가 끝나면, 현재 인덱스와 가장 작은 숫자 인덱스의 숫자를 자리바꿈

## 예제 21. 선택 정렬 구현

```python
def selection_sort(nums):
    n = len(nums)
    # 모든 원소를 순차적으로 순회
    for i in range(n):
        # 현재 인덱스를 가장 작은 숫자로 설정하고
        smallest_idx = i
        # 현재 인덱스 이후의 원소를 돌면서
        for j in range(i + 1, n):
            # 현재 인덱스 숫자보다 작으면 해당 숫자를 가장 작은 숫자로 설정
            if nums[i] > nums[j]:
                smallest_idx = j
        # 한 번의 순회마다 한 번의 교환만 진행
        nums[i], nums[smallest_idx] = nums[smallest_idx], nums[i]
        print(nums)
    return nums

# 예시로, 무작위로 배치된 1부터 10까지의 숫자 리스트가 있다고 가정해봅시다.
random_nums = [3, 1, 4, 5, 2, 10, 9, 6, 8, 7]
selection_sort(random_nums)
```
