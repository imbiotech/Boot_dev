# 07. Data Structures

이번에는 더 나은 알고리즘을 이용해 실제로 자료가 어떻게 효율적으로 저장 및 정렬되는지 알아볼까 합니다. `자료 구조`는 좀 더 나은 알고리즘을 가진 도구입니다. 예를 들어 다음과 같은 것들이 있습니다.

- Stack: 후입 선출형 자료 구조 (Last In First Out, LIFO)
- Queues: 선입 선출형 자료 구조 (First In First Out, FIFO)
- Linked List: Node의 연쇄, 삽입과 삭제에 효율적인 자료 구조
- Binary Tree: 각 Node가 두 개의 자식 Node를 가지는 트리 구조
- Red Black Tree: 색깔을 활용해 자체적으로 양 쪽 밸런스를 잡아주는 트리 구조
- Hashmaps: 키와 값을 가지는 자료 구조
- Tries: 단어의 저장과 검색을 빠르게 해주는 트리 구조
- Graphs: Edge로 이어진 Node의 모음

## 예제 25. 리스트에서 원하는 데이터의 개수를 찾는 알고리즘

- 문자열의 리스트를 받아 리스트 내에서 내가 원하는 문자열이 포함되었는 지를 확인하는 알고리즘을 만들어 봅시다. 예제에서는 리스트 내에서 `‘marketer’`를 세고 그 숫자를 반환합니다.

````python
def count_marketers(job_titles):
	"""
	우선 가장 간단하게는 

		return job_titles.count("marketer")
	
	이렇게 처리할 수 있으나,
	이 방법의 가장 큰 문제는 MARKETER 같이 대문자가 포함된
	글자를 셀 수 없습니다. 
	그러므로 각 문자를 소문자로 전환하여 비교하는 것이 정확합니다.
	"""
	
	count = 0
	
	for job in job_titles:
			if job.lower() == "marketer":
					count += 1

	return count

	"""
	또는 아래와 같이 구성할 수도 있고요
	"""
	return sum(
		1
		for job in job_titles
		if job.lower() == "marketer"
	)

# 테스트 코드
job_titles = ["Marketer", "marketer", "MARKETER", "Developer", "Designer"]
print(count_marketers(job_titles))  # 출력: 3
````

## 리스트 (List)

우선 리스트를 한 번 살펴봅시다. 리스트를 자료 구조나 알고리즘 관점으로 볼 때 어떤 것이 좋을까요? 리스트는 다음과 같은 내장 함수를 동반합니다.

- Append: 리스트 가장 마지막에 데이터를 추가할 때 사용합니다. 얼마나 많이 추가 하는지에 무관하게 들어온 순서대로 가장 뒤쪽에 계속 배치하기 때문에 시간 복잡도는 $O(1)$입니다.
    
    ```python
    animals = ["cat", "dog", "mouse"]
    
    # 이와 같은 리스트가 있을때 리스트에 cow를 추가해봅시다.
    
    animals.append("cow")
    
    # 이제 리스트는 ["cat", "dog", "mouse", "cow"]가 됩니다.
    ```
    
- Index: 특정 위치에 있는 요소를 읽어올 때 사용합니다. 마찬가지로 얼마나 많은 원소를 가진 리스트에서 호출하는지는 무관하게 항상 시작부터 n 번째 위치에 있는 원소를 가져오므로 시간 복잡도는 $O(1)$입니다.
    
    ```python
    print(animals[1])
    
    # 이 코드는 animals 리스트에 자료가 100 개든,
    # 100 만 개 든# 항상 두 번째 원소를 가져옵니다.
    # 이 코드에서는 dog이 출력됩니다.
    ```
    
- Delete: 특정 위치에 있는 요소를 삭제할 때 사용합니다. 리스트는 중간이 비어있을 수 없기 때문에, 중간에 있는 원소를 삭제하고 나면 남은 모든 원소를 앞으로 당겨야 합니다. 이러한 과정은 매 원소마다 반복적으로 진행되므로, 시간 복잡도는 $O(n)$입니다.
    
    ```python
    animals.pop(1)
    
    # 이 코드는 리스트의 두 번째 인덱스에 있는 요소를
    # 삭제 및 삭제된 내용을 출력합니다.
    # 즉, 위 리스트에서 "dog"이 삭제 및 출력되고,
    # 나머지 요소를 앞으로 당겨
    # ["cat", "mouse", "cow"]가 됩니다.
    ```
    
- Search: 어떤 원소가 리스트에 있는지를 확인하고, 그 위치를 알고 싶을 때 사용합니다. 리스트의 각 원소가 내가 찾는 값인지를 비교하고 결과를 출력하기 때문에 시간 복잡도는 역시 $O(n)$입니다.
    
    ```python
    animals.index("cow")
    
    # 이 코드는 리스트 내에 cow가 있는지 확인하고,
    # cow가 존재하면 인덱스를, 없으면 오류를 반환합니다.
    # 현재 리스트가 ["cat", "mouse", "cow"]이므로
    # 반환값은 2가 됩니다.
    ```
    

정리하자면, 리스트는 두 가지 기본적인 로직, Delete와 Search에서 불리한 알고리즘입니다.

## 예제 26. 리스트의 가장 마지막 데이터를 받아오기

- (가장 과거부터 가장 최신까지 순차적으로 정렬된) 리스트를 입력으로 받아, 가장 최근의 데이터를 반환하는 알고리즘을 만들어봅시다.

````python
def last_work_experience(work_experiences):
    # 리스트가 비어있으면 None을 반환합니다.
    if not work_experiences: 
        return None
    
    # 그 외에는 리스트의 가장 마지막 인덱스를 반환합니다.
    # 파이썬에서는 리스트의 길이를 모를 때 가장 마지막 인덱스를 출력하기 위해 -1을 인덱스로 사용합니다.
    return work_experiences[-1]

    # 또는 한 줄로 표현하면 이렇게 됩니다.
    return work_experiences[-1] if len(work_experiences) > 0 else None
		
# 테스트 코드
work_experiences = ["Company A", "Company B", "Company C"]
print(last_work_experience(work_experiences))  # 출력: "Company C"
````

## 스택 (Stacks)

스택은 리스트와 비슷하게 자료를 순차적으로 저장하는 구조입니다. 다만, 좀 더 제한적입니다. `쌓다`라는 의미에 걸맞게 스택의 자료 구조에서는 모든 데이터가 가장 꼭대기에만 추가되거나, 꼭대기에서만 제거될 수 있습니다. 큰 택배 박스를 순서대로 차곡차곡 쌓는다고 생각해봅시다. 갑자기 가장 아래나 중간 즈음에 택배 박스를 추가하고 싶다고 마음대로 추가할 수 없죠. 그렇기 때문에 스택은 앞서 설명한
`후입 선출 (Last In First Out, LIFO)`의 특성을 가집니다. 그럼 도대체 이런 귀찮은 녀석을 어디다 쓰죠…?

## 예제 27. 스택 구현 및  push, size 메서드 구현

- 스택은 매우 유용하게 쓰입니다. 우리가 엑셀이나 워드를 만질 때 너무나 일상적으로 당연하게 사용하는 기능 중 하나인 되돌리기 (ctrl+z)와 다시 실행 (ctrl+y)는 잘 생각해보면 스택 구조의 일종입니다.
- 스택의 기능 중 push와 size를 먼저 구현해봅시다. 앞서 언급한 것처럼 스택은 리스트와 유사하므로 리스트를 약간 개조해서 스택을 구현할 겁니다.
    - push: 스택의 최상단에 요소를 추가하는 기능입니다. 리스트와 비교하자면, 리스트의 가장 마지막이 스택의 최상단이 되겠죠.
    - size: 스택의 사이즈를 반환합니다.
- 여기서부터는 class 기능을 적극 활용합니다. 위에서 잠깐 나온 적이 있지만 다루진 않았습니다. class는 프로그래밍 언어에서 비슷한 기능을 반복적으로 수행할 때 쉽게 호출하고, 코드를 중복해서 작성하는 것을 방지하기 위한 하나의 묶음이라고 보면 됩니다. 예를 들어, list 역시 파이썬에 내장된 클래스의 하나로, list.append()나 list.pop()등의 기능을 손쉽게 쓸 수 있는 것도 클래스 단위에서 기능이 구축되었기 때문입니다.

````python
class Stack:
    # __init__ 메서드는 Stack 객체가 생성될 때 
    # 호출됩니다. 빈 리스트를 하나 생성해 
    # 스택에 저장합니다.
    def __init__(self):
        self.items = []

    # push 메서드는 스택의 최상단 (리스트의 맨 끝)에 
    # 새로운 데이터를 추가합니다.
    def push(self, item):
        self.items.append(item)

    # size 메서드는 스택 내에 있는 데이터의 개수를 
    # 반환합니다.
    def size(self):
        return len(self.items)

# 테스트 코드
stack = Stack()
stack.push(1)
print(stack.items)  # 출력: [1]
stack.push(2)
print(stack.items)  # 출력: [1, 2]
print(stack.size())  # 출력: 2
````

## 스택의 시간 복잡도

스택에 대해서 배웠으니 아마 이런 생각이 드실 수 있습니다. “그냥 좀 더 불편한 리스트 같은데? 왜 이게 필요하지?” 네, 그 생각이 맞습니다. 스택은 좀 더 기능이 적은 리스트고, 그게 바로 포인트입니다. 기능을 제한하는 방식으로 매우 빠른 알고리즘을 만든거죠. 위에서 배운 push와 size를 포함해 스택에는 총 4 가지의 메서드가 존재합니다.

- Push: 스택의 가장 상단에 새로운 데이터를 추가합니다. 추가하는 위치가 확실하기 때문에, 많은 데이터를 추가하더라도 항상 시간 복잡도는 $O(1)$입니다.
- Pop: 스택의 가장 상단에 있는 데이터를 제거하고 해당 값을 반환합니다. 마찬가지로 제거하는 위치가 확실하기 때문에, 많은 데이터를 제거하더라도 항상 시간 복잡도는 $O(1)$입니다.
- Peek: 스택의 가장 상단에 있는 데이터를 반환합니다. Pop과 달리 삭제하지 않으며, 역시 시간 복잡도는 항상 $O(1)$입니다.
- Size: 스택의 데이터 개수를 반환합니다. 개수가 몇 개든 항상 시간 복잡도는 $O(1)$입니다.

이처럼 모든 메서드가 $O(1)$의 시간 복잡도를 가지기 때문에 스택 자체의 사이즈가 얼마나 되든 항상 일정한 시간이 걸립니다.

## 예제 28. 스택의 peek, pop 메서드 구현

- 나머지 메서드도 구현해봅시다.

````python
class Stack:
    # __init__ 메서드는 Stack 객체가 생성될 때 호출됩니다.
    # 빈 리스트를 하나 생성해 스택에 저장합니다.
    def __init__(self):
        self.items = []

    # push 메서드는 스택의 최상단 (리스트의 맨 끝)에 새로운 데이터를 추가합니다.
    def push(self, item):
        self.items.append(item)

    # size 메서드는 스택 내에 있는 데이터의 개수를 반환합니다.
    def size(self):
        return len(self.items)

    # peek 메서드는 스택의 최상단에 있는 데이터를 반환합니다.
    def peek(self):
        if not self.items:
            return None
        return self.items[-1]

    # pop 메서드는 스택의 최상단에 있는 데이터를 제거하고 반환합니다.
    def pop(self):
        if not self.items:
            return None
        return self.items.pop()
    
# 테스트 코드
stack = Stack()
stack.push(1)
print(stack.items)  # 출력: [1]
stack.push(2)
print(stack.items)  # 출력: [1, 2]

print(stack.size())  # 출력: 2
print(stack.peek())  # 출력: 2
print(stack.items)  # 출력: [1, 2]
print(stack.pop())   # 출력: 2
print(stack.items)  # 출력: [1]
````

## 스택 정리

- 모든 메서드 알고리즘의 시간 복잡도가 $O(1)$이지만, 실제로 가장 아래에 있는 데이터에 접근하고 싶다면, 더 높은 시간 복잡도가 발생합니다.
- 스택의 알고리즘에서는 탐색과 정렬, 무작위 접근이 불가능합니다.
- 스택은 다른 추상 데이터 타입과 동일하게 어떠한 데이터 형식이라도 스택 형태로 저장할 수 있습니다.
- 스택은 실무에서 `함수 호출 관리`, `실행 취소/재실행 기능`, `수식 계산`, `웹브라우저 기록` 등에서 사용됩니다.

## 예제 29. 스택을 활용한 괄호 검사

- 엑셀에서 조금 복잡한 수식이나 IF 조건문을 사용하는 경우를 생각해봅시다. 이 때 여는 괄호와 닫는 괄호가 동일한 개수만큼 들어있지 않으면 엑셀은 오류를 반환합니다. 이건 어떤 원리로 동작하는 걸까요?
- 예를 들어,
    - `()`
    - `()()`
    - `((()))`
    - `(()(()))`
    
    이 괄호는 균형이 맞는 괄호지만,
    
    - `(`
    - `())`
    - `(()()`
    - `(()))`
    
    이 괄호는 균형이 맞지 않습니다.
    
- 스택을 활용해 괄호의 균형이 맞는지 여부를 반환하는 코드를 짜봅시다. (앞서 구현한 Stack Class를 활용합니다.)

````python
def is_balanced(input_str):
    # 새 Stack 생성
    stack = Stack()

    # 입력된 문자열을 순차적으로 스택에 쌓습니다.
    for t in input_str:
        stack.push(t)
    
    # 스택을 가장 위쪽부터 순차적으로 출력하면서 괄호의 밸런스가 맞는지를 검증합니다.
    count = 0
    while stack.size() > 0:
        # 닫는 괄호는 +1, 여는 괄호는 -1을 부여합니다.
        if stack.pop() == ")":
            count += 1
        else:
            count -= 1
        
        # 만약 수치가 음수로 내려간다면, 닫는 괄호 없이 여는 괄호가 먼저 있다는 것을 뜻하고, 
        # 이미 밸런스 맞지 않습니다. 거짓을 반환합니다.
        if count < 0:
            return False
    
    # 수치가 0보다 크면 밸런스가 맞지 않다는 의미입니다.
    if count > 0:
        return False
    return True

# 테스트 코드
print(is_balanced("()()"))  # 출력: True
print(is_balanced("(()())"))  # 출력: True
print(is_balanced("(()"))  # 출력: False
print(is_balanced("())"))  # 출력: False
````

## 큐 (Queue)

- 큐는 순서대로 저장된 원소들… 맞습니다. 리스트랑 또 비슷한 거죠? 스택과 마찬가지로 약간 제한적인 구조를 가지고 있는 자료 구조입니다. 예를 들어, 식당이나 카페에서 주문을 처리하는 방법, 롤 같은 다인 매칭 시스템에서 게임 대기열을 처리하는 방법 등이 큐입니다. `선입선출 (First In First Out, FIFO)`의 특성을 가집니다.

- 큐는 다음 메서드를 가지고 있습니다.
    - `queue.push(item)`: Queue의 마지막 (List를 기준으로 가장 처음)에 원소를 추가합니다. 언어에 따라 `enqueue`라고도 합니다.
    - `queue.pop()`: Queue의 처음 원소 (List를 기준으로 가장 마지막)를 삭제하고 반환합니다. 언어에 따라 `dequeue`라고도 합니다.
    - `queue.peek()`: Queue의 처음 원소를 반환합니다. (삭제하지 않습니다.)
    - `queue.size()`: Queue의 원소 개수를 반환합니다.

## 예제 30. 큐 구현

- 큐의 4가지 메서드를 구현해봅시다.

````python
class Queue:
    def __init__(self):
        self.items = []

    # 큐의 가장 마지막에 아이템을 추가합니다.
    def push(self, item):
        self.items = [item] + self.items

    # 큐의 사이즈가 0보다 클 때 가장 첫 원소를 꺼냅니다.
    def pop(self):
        return None if self.size() <= 0 else self.items.pop()

    # 큐의 사이즈가 0보다 클 때 가장 첫 원소를 보여줍니다.
    def peek(self):
        return None if self.size() <= 0 else self.items[-1]

    # 큐의 사이즈를 반환합니다.
    def size(self):
        return len(self.items)
    
# 테스트 코드
queue = Queue()
queue.push(1)
print(queue.items)  # 출력: [1]
queue.push(2)
print(queue.items)  # 출력: [2, 1]
print("=="*20)
print(queue.size())  # 출력: 2
print("=="*20)
print(queue.peek())  # 출력: 1
print(queue.items)  # 출력: [2, 1]
print("=="*20)
print(queue.pop())   # 출력: 1
print(queue.items)  # 출력: [2]
````

## 큐의 시간 복잡도

큐의 기본 원리는 스택과 거의 유사합니다. 즉, 시간 복잡도 역시 동일하다는 것을 의미합니다.

- Push: 큐의 가장 마지막에 새로운 데이터를 추가합니다. 추가하는 위치가 확실하지만, 리스트의 가장 앞에 새로운 데이터를 추가하는 방식의 특성으로 인해 새 요소 뒤에 기존 리스트를 복사하는 방식으로 구현됩니다. 그러므로 시간 복잡도는 $O(n)$이 됩니다.
- Pop: 큐의 가장 처음에 있는 데이터를 제거하고 해당 값을 반환합니다. 마찬가지로 제거하는 위치가 확실하기 때문에, 많은 데이터를 제거하더라도 항상 시간 복잡도는 $O(1)$입니다.
- Peek: 큐의 가장 처음에 있는 데이터를 반환합니다. Pop과 달리 삭제하지 않으며, 역시 시간 복잡도는 항상 $O(1)$입니다.
- Size: 큐의 데이터 개수를 반환합니다. 개수가 몇 개든 항상 시간 복잡도는 $O(1)$입니다.

다만, 큐의 인덱스가 리스트의 인덱스와 반대로 동작하기 때문에 약간 헷갈리실 수는 있습니다.

## 예제 31. 매치 메이킹 (Matchmaking)

- 위에서 큐의 예시로 롤을 예시로 들었는데, 롤의 매치 메이킹 시스템을 한 번 구현해봅시다.
- 이번 예제를 위해서 실제적으로 큐의 규칙을 따르지는 않지만 현업에서의 문제 해결에 도움이 되는 메서드인 `search_and_remove`를 하나 추가하여 사용합니다.
- 항상 규칙이 절대적일 필요는 없습니다. 필요에 따라서 기존에 마련된 규칙에 어긋나는 것을 사용할 필요도 있습니다. 다만, 그 때 발생할 수 있는 장단점에 대해서 잘 이해해야 합니다.

````python
class Queue:
    def __init__(self):
        self.items = []

    # 큐의 가장 마지막에 아이템을 추가합니다.
    def push(self, item):
        self.items = [item] + self.items

    # 큐의 사이즈가 0보다 클 때 가장 첫 원소를 꺼냅니다.
    def pop(self):
        return None if self.size() <= 0 else self.items.pop()

    # 큐의 사이즈가 0보다 클 때 가장 첫 원소를 보여줍니다.
    def peek(self):
        return None if self.size() <= 0 else self.items[-1]

    # 큐의 사이즈를 반환합니다.
    def size(self):
        return len(self.items)

    # 해당 원소가 리스트에 있다면
    # 원소의 `인덱스에 무관하게` 삭제하고
    # 해당 원소를 반환합니다.
    def search_and_remove(self, item):
        if item not in self.items:
            return None
        self.items.remove(item)
        return item

# 테스트 코드
queue = Queue()
queue.push(1)
print(queue.items)  # 출력: [1]
queue.push(2)
print(queue.items)  # 출력: [2, 1]
print("=="*20)
print(queue.search_and_remove(1))  # 출력: 1
print(queue.items)  # 출력: [2]
````

- 매치 메이킹 시스템은 다음 형식을 따릅니다.
    - `떠나기`에 대해 큐에서 유저를 찾아 큐에 유저가 있으면 삭제합니다.
    - `참여하기`에 대해 큐에 유저를 올립니다.
    - 적어도 4명의 유저가 모이면, 그 중 첫 두 명을 큐에서 제거하고, `유저1과 유저2가 매칭되었습니다!(user1 mathched user2!)`라는 문자열을 반환합니다.
    - 4명 미만의 유저가 큐에 있을 경우 `매치되지 않았습니다(No match found)`라는 문자열을 반환합니다.

````python
def matchmake(queue, user):
    print(f"Queue items: {queue.items}")
    if (user[-1] == "leave") and (user[0] in queue.items):
        queue.search_and_remove(user[0])
    elif (user[-1] == "join") and (not(user[0] in queue.items)):
        queue.push(user[0])
    if queue.size() >= 4:
        user1 = queue.pop()
        user2 = queue.pop()
        return f"{user1} matched {user2}!"
    else:
        return "No match found"
    return queue

# 테스트 코드
queue = Queue()
print(matchmake(queue, ["Alice", "join"]))  # 출력: "No match found"
print(queue.items)  # 출력: ['Alice']
print("=="*20)
print(matchmake(queue, ["Bob", "join"]))    # 출력: "No match found"
print(queue.items)  # 출력: ['Bob', 'Alice']
print("=="*20)
print(matchmake(queue, ["Charlie", "join"]))  # 출력: "No match found"
print(queue.items)  # 출력: ['Charlie', 'Bob', 'Alice']
print("=="*20)
print(matchmake(queue, ["Dave", "join"]))  # 출력: "Alice matched Bob!"
print(queue.items)  # 출력: ['Charlie', 'Dave']
print("=="*20)
print(matchmake(queue, ["Charlie", "leave"]))  # 출력: "No match found
print(queue.items)  # 출력: ['Dave']
print("=="*20)
````

## 연결 리스트 (Linked List)

Queue의 push의 시간 복잡도는 $O(n)$이기 때문에 이를 해결할 좀 더 개선된 형태의 Queue가 필요합니다. 이를 위해선 기존의 리스트 대신 `연결 리스트(Linked List)`라는 새로운 형태를 사용합니다.

연결 리스트는 기존의 리스트처럼 한 원소의 바로 옆 메모리 위치에 새로운 원소를 저장하는 방식이 아니라, 각각의 원소가 자신의 데이터와 함께 다음 원소의 레퍼런스를 가지고 있는 방식입니다.

## `연결 리스트`와 `리스트` 비교

연결 리스트는 리스트와 비슷하게 순서대로 정렬된 자료의 모음 형태입니다. 일반적인 리스트의 원소는 메모리 상에서 바로 옆에 위치하기 때문에 숫자로 지정된 인덱스를 가질 수 있습니다. 이렇게 주어진 인덱스는 다르게 말하면 시작 지점으로부터 얼마나 떨어져있는지에 대한 값이라고도 볼 수 있죠. 

---
```python
car = cars[3]
```
---

반면, 연결 리스트는 인덱스가 없습니다. 데이터 값과 다음 데이터에 대한 위치만 알려줄 뿐입니다. 때문에 연결 리스트에 대해서 어떤 반복 작업을 하기 위해서는 첫 노드부터 시작해서 계속 각 노드의 next를 찾아가는 방식이 됩니다.

---
```python
current_car_node = head_car_node
while current_car_node is not None:
	print(current_car_node.val)
	current_car_node = current_car_node.next
```
---

솔직히 연결 리스트는 구현 방식과 사용 방식에 있어서 복잡하고 부하를 더 일으킬 것 같은데, 왜 쓰는 걸까요? 그건 연결 리스트가 삽입과 삭제에 대해 매우 빠르게 동작하기 때문입니다. 일반적인 리스트에서는 앞서 언급했던 것처럼, 삽입과 삭제에 대한 시간 복잡도는 $O(n)$이라고 했습니다.

반면, 연결 리스트에서는 중간에 값을 끼워 넣기 위해서 바로 앞 노드의 next를 새로운 노드로, 새로운 노드의 next를 다음 노드로 이어주기만 하면 되기 때문에 시간 복잡도가 $O(1)$로 향상됩니다.

## 예제 33. 연결 리스트의 반복 작업 (iterating) 구현

- 연결 리스트의 반복 작업이 리스트 같은 일반 배열보다 상당히 안 좋더라도 우리는 공부 중이니 일단 구현해봅시다.
- Python에서는 `__iter__` 메서드가 구현되어 있어 좀 더 쉽게 구현할 수 있습니다. 또한, `yield`라는 키워드를 사용해야 합니다. `yield`는 `return`과 비슷하지만, `함수(function)`를 `생성 함수(generator function)`로 바꿔줍니다.
    - 생성 함수 (Generator function)
    
        생성 함수는 함수와 비슷하게 내부 코드를 실행하는 형태로 구성됩니다. 단, return을 만나면 종료되는 함수와 달리 yield를 만나면 멈춥니다. 그 이후 두 번째로 해당 함수가 호출되면, yield 이후를 실행합니다.
        
        

````python
def create_message_generator():
    yield "hi"
    yield "there"
    yield "friend"

gen = create_message_generator()

# 처음 함수를 호출하면
first = next(gen)
print(first)
# *hi*가 출력되고, 해당 부분에 함수는 멈춰있습니다.

# 두 번째로 함수를 호출하면
second = next(gen)
print(second)  
# *there*이 출력되고, 해당 부분에 함수는 멈춰있습니다.

# 세 번째로 함수를 호출하면
third = next(gen)
print(third)  
# *friend*가 출력되고 함수는 종료됩니다.

# 그럼 만약 네 번째로 함수를 호출하면?
try:
    fourth = next(gen)
except StopIteration as e:
    print("No more messages to generate.")
    # **StopIteration Exception**이 발생합니다.

gen2 = create_message_generator()

# 반복문으로 호출합니다.
# 반복문은 내부에 몇 개의 yield가 있는지 신경쓰지 않고
# 호출해서 사용할 수 있다는 장점이 있습니다.
for message in gen2:
    print(message)
````

단, 매번 `create_message_generator()`를 호출하면 새로운 인스턴스를 생성하기 때문에 매번 `iteration`도 초기화 됩니다. 그러므로 gen에 최초로 생성한 generator를 할당한 후 반복해서 호출하는 방식을 사용합니다.
        
그 후 `next()` 메서드를 사용하거나, 반복문에서 호출하는 방식으로 해결합니다.

````python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

    def set_next(self, node):
        self.next = node

    def __repr__(self):
        return self.val


class LinkedList:
    # 최초의 연결 리스트는 빈 head 자리만 가진 상태로 시작합니다.
    def __init__(self):
        self.head = None

    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next

    # don't touch below this line

    def __repr__(self):
        nodes = []
        current = self.head
        while current and hasattr(current, "val"):
            nodes.append(current.val)
            current = current.next
        return " -> ".join(nodes)
    
# 테스트 코드
linked_list = LinkedList()
linked_list.head = Node("Major Marquis Warren")
print(linked_list)  # 출력: Major Marquis Warren
print("=="*20)
linked_list.head.set_next(Node("John Ruth"))
print(linked_list)  # 출력: Major Marquis Warren -> John Ruth
print("=="*20)
linked_list.head.next.set_next(Node("Django"))
print(linked_list)  # 출력: Major Marquis Warren -> John Ruth -> Django
print("=="*20)
for node in linked_list:
    print(f"node {node} is in the linked list. And its next node is {node.next}.")
````

## 예제 34. 연결 리스트의 마지막에 노드 추가하기 (Add to tail)
- 이번에는 이미 존재하는 연결 리스트의 가장 마지막에 새로운 노드를 추가해봅시다. 파이썬 리스트의 append 메서드 같은 기능요.

<img src="https://storage.googleapis.com/qvault-webapp-dynamic-assets/course_assets/7UXPMpU-1280x383.png">

- add_to_tail 메서드는 다음과 같은 방식을 따릅니다.
    1. head가 없다면, 새로운 노드를 head로 설정하고 마칩니다.
    2. heed가 있다면, 연결 리스트의 가장 마지막 노드를 찾습니다.
        - last_node를 None으로 선언한 뒤 `__iter__` 메서드를 사용해서 iterating 합니다.
    3. 전체 리스트를 순환했다면, 마지막 노드에 도달했을 것이고, 해당 노드의 next를 새로운 노드로 설정합니다.

````python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

    def set_next(self, node):
        self.next = node

    def __repr__(self):
        return self.val

class LinkedList:
    def add_to_tail(self, node):
        if self.head == None:
            self.head = node
            return
        last_node = None
        for my_node in self:
            last_node = my_node
        last_node.set_next(node)

    # don't touch below this line

    def __init__(self):
        self.head = None

    def __iter__(self):
        node = self.head
        while node is not None:
            yield node
            node = node.next

    def __repr__(self):
        nodes = []
        for node in self:
            nodes.append(node.val)
        return " -> ".join(nodes)
    
# 테스트 코드
linked_list = LinkedList()
linked_list.add_to_tail(Node("Major Marquis Warren"))
linked_list.add_to_tail(Node("John Ruth"))
linked_list.add_to_tail(Node("Django"))
print(linked_list)  # 출력: Major Marquis Warren -> John Ruth -> Django
````

## 예제 35. 연결 리스트의 처음에 노드 추가하기 (Add to head)
- 이번에는 이미 존재하는 연결 리스트의 가장 처음에 새로운 노드를 추가해봅시다. 파이썬 리스트의 insert 메서드 같은 기능요.
1. 새로운 노드의 set_next를 기존 head로 설정합니다.
2. head를 새로운 노드로 설정합니다.

````python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

    def set_next(self, node):
        self.next = node

    def __repr__(self):
        return self.val

class LinkedList:
    def add_to_head(self, node):
        node.set_next(self.head)
        self.head = node

    # don't touch below this line

    def add_to_tail(self, node):
        if self.head is None:
            self.head = node
            return
        last_node = None
        for current_node in self:
            last_node = current_node
        last_node.set_next(node)

    def __init__(self):
        self.head = None

    def __iter__(self):
        node = self.head
        while node is not None:
            yield node
            node = node.next

    def __repr__(self):
        nodes = []
        for node in self:
            nodes.append(node.val)
        return " -> ".join(nodes)
    
# 테스트 코드
linked_list = LinkedList()
linked_list.add_to_head(Node("Django"))
linked_list.add_to_head(Node("John Ruth"))
print(linked_list)  # 출력: John Ruth -> Django
````
