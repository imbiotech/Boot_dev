- List: 순서가 지정된 데이터 묶음
- Dictionary: 키와 값을 매칭시켜 저장한 데이터
- Sets: 순서가 지정되지 않은 고유 값 데이터

알고리즘 (Algorithm): 어떤 문제를 해결하기 위해 수행하는 지시의 모음

## 예제 1. 최소값 찾기

- 숫자형 자료가 들어있는 List를 입력받아 그 중 가장 작은 수를 출력하는 알고리즘

```python
def find_minimum(nums):
		if len(nums) <= 0: # 입력받은 데이터가 없을 때
				return None # None (없음)을 반환
		
		minimum = float("inf") # minimum의 초기값을 무한대로 설정

		for num in nums: # nums 안에 존재하는 숫자를 하나씩 순환
				if minimum > num: # 만약 minimum이 num보다 크다면, 현재까지 num이 최소값임
						minimum = num # minimum을 num으로 치환

		return minimum # 최종 minimum을 반환

find_minimum([3, 1, 4, 1, 5, 9]) # 예시로 리스트를 입력하여 최소값을 찾음
```

```python
def find_minimum(nums):
		if len(nums) <= 0: # 입력받은 데이터가 없을 때
				return None # None (없음)을 반환

		return min(nums) # List형 데이터에서 최소값을 출력하는 python 기본 함수

find_minimum([3, 1, 4, 1, 5, 9]) # 예시로 리스트를 입력하여 최소값을 찾음
```

## 알고리즘이란?

컴퓨터 사이언스 영역에서 알고리즘은 “명확하게 정의되고, 컴퓨터가 실행가능한 유한한 숫자의 지시”를 의미함. 어떠한 문제를 해결하는 개략적인 해결책이기에 종종 의사 코드 (pseudocode)로 기술하기도 함. 알고리즘 자체는 Python, Go, TypeScript 등등 어떠한 프로그래밍 언어로 구현하는지는 무관함.


## 예제 2. 어떤 알고리즘이 하는 역할 찾기

다음과 같은 역할을 하는 알고리즘이 있다면, 이 알고리즘은 무엇을 하는가?

- 어떤 문자형 데이터가 들어있는 변수 S가 있고, 비어있는 문자형 변수 R이 있다. 변수 S의 문자를 하나씩 순환하며 R에 차곡차곡 앞에서부터 쌓는다. 변수 S의 모든 문자를 순환하고 나면 결과값 R을 반환한다.

 ⇒ 해당 알고리즘은 글자를 뒤집는 역할을 한다.

## 간단한 알고리즘

알고리즘은 문제를 어렵게 만들어보이는 단어지만, 실제로는 일상 생활에서 아주 쉽게 쓰입니다. 예를 들어 두 숫자 a, b를 입력 받아, 그 두 숫자를 + 기호를 사용해 더한 뒤 결과값인 합계, sum을 반환하는 간단한 과정도 일종의 알고리즘입니다.

## 예제3. 합계

- 숫자형 자료가 들어있는 List를 입력받아 합계를 출력하는 알고리즘

```python
def summed(nums):
    summed = 0 # 합산에 사용하기 위한 변수를 선언
    for num in nums:
        summed += num # 모든 데이터를 순환하며 summed에 합산
				
    return summed

summed([1, 2, 3, 4, 5]) # 예시로 리스트를 입력하여 합산값을 찾음
```

```python
def summed(nums):
    return sum(nums) # List형 데이터에서 합산값을 출력하는 python 기본 함수

summed([1, 2, 3, 4, 5]) # 예시로 리스트를 입력하여 합산값을 찾음
```

## 예제 4. 확산 (Spread)

Social Media 업계에는 확산 (Spread)라는 개념이 있습니다. 원글 작성자의 팔로워에 의해서 글이 얼마나 빠르게 퍼져나가는지에 대한 예측으로, 기하급수적으로 상승하는 것으로 확인됐습니다.

$$
예측\;확산\;수치\;(\text{estimated\_spread}) = \\
원글\;작성자의\;팔로워들의\;평균\;팔로워수\;(\text{average\_audience\_followers})\;*\;(원글\;작성자의\;팔로워수\;(\text{num\_followers})\; \^{} \;1.2)
$$

- 원글 작성자의 각 팔로워들의 팔로워수를 리스트형으로 받아, Spread를 계산하는 알고리즘

```python
def get_estimated_spread(audiences_followers):
    if len(audiences_followers) <= 0:
        return 0

    num_followers = 0
    average_audience_followers = 0

    for audience_follower in audiences_followers:
        num_followers += 1
        average_audience_followers += audience_follower

    average_audience_followers = average_audience_followers / num_followers

    estimated_spread = average_audience_followers * (num_followers ** 1.2)

    return estimated_spread

get_estimated_spread([100, 200, 300]) # 예시로 리스트를 입력하여 Spread 수치를 찾음
```

```python
def get_estimated_spread(audiences_followers):
    if len(audiences_followers) <= 0:
        return 0

    num_followers = len(audiences_followers)
    average_audience_followers = sum(audiences_followers)/num_followers
    estimated_spread = average_audience_followers * (num_followers ** 1.2)

    return estimated_spread

get_estimated_spread([100, 200, 300]) # 예시로 리스트를 입력하여 Spread 수치를 찾음
```
