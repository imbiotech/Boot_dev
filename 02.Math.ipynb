## 예제 5. 팔로워 추이 예측

- 인플루언서의 현재 팔로워 수 (정수형), 활동 분야 (문자형)를 가지고 몇 개월 (정수형) 뒤에 팔로워 수가 어떻게 될지 예측하는 알고리즘
- 활동 분야는 “fitness”, “cosmetic”, “other”이 있으며, 분야에 따라 각각 매달 4배, 3배, 2배 씩 증가
- 이렇게 어떤 숫자에 특정 숫자를 계속 곱해서 규칙성을 만들어 나가는 방법을 기하 수열 또는 기하 급수라고 하며, 흔히 알고 있는 등비 수열도 이러한 기하 수열의 예시

```python
def get_follower_prediction(follower_count, influencer_type, num_months):
    a1 = follower_count
    
    if influencer_type == "fitness":
        r = 4
    elif influencer_type == "cosmetic":
        r = 3
    else:
        r = 2

    return a1 * (r ** (num_months))

get_follower_prediction(1000, "fitness", 6) # 예시로 팔로워 수, 인플루언서 유형, 개월 수를 입력하여 예측값을 찾음
```

```python
def get_follower_prediction(follower_count, influencer_type, num_months):
    a1 = follower_count
    
    r = 4 if influencer_type == "fitness" else 3 if influencer_type =="cosmetic" else 2

    return a1 * (r ** (num_months))

get_follower_prediction(1000, "fitness", 6) # 예시로 팔로워 수, 인플루언서 유형, 개월 수를 입력하여 예측값을 찾음
```

## 로그 (Log)

로그는 지수의 반대 개념입니다. 파이썬에서 로그를 사용하기 위해서는 math라는 라이브러리를 가져와야 하며 `import math`를 통해 라이브러리를 불러온 후 `math.log()`를 입력하여 사용할 수 있습니다.

## 예제 6. 인플루언서 점수

- 인플루언서의 평균 참여율에 팔로워 수의 log(2) 값을 곱한 결과를 도출하는 알고리즘

```python
import math

def get_influencer_score(num_followers, average_engagement_percentage):
    return average_engagement_percentage * math.log(num_followers, 2)

get_influencer_score(10000, 5) # 예시로 팔로워 수와 평균 참여율을 입력하여 인플루언서 점수를 찾음
```

## 계승 (Factorial)

어떤 정수의 계승은 1부터 해당 정수까지를 모두 곱한 값입니다. $n! = n \times (n-1) \times ... \times 2 \times 1$로 계산합니다.

## 예제 7. 순서 찾기

- 인플루언서가 게시 예정인 포스팅의 수를 토대로, 가능한 모든 순서를 계산하는 알고리즘

```python
def num_possible_orders(num_posts):
    result = 1
    for i in range(1, num_posts + 1):
        result *= i
    return result

num_possible_orders(5) # 예시로 게시물 수를 입력하여 가능한 순서의 수를 찾음
```

```python
import math

def num_possible_orders(num_posts):
    return math.prod([i for i in range(1, num_posts + 1)])

num_possible_orders(5) # 예시로 게시물 수를 입력하여 가능한 순서의 수를 찾음
```

## 지수 감소 (Exponential Decay)

현재값 대비 어떤 비율의 지수 만큼 감소하는 방식

## 예제 8. 팔로워 감소

- 인플루언서가 활동을 하지 않을 때 매일 줄어드는 팔로워 수를 계산하는 알고리즘

```python
def decayed_followers(initial_followers, fraction_lost_daily, days):
    return initial_followers * (1-fraction_lost_daily)**days

decayed_followers(1000, 0.1, 30) # 예시로 초기 팔로워 수, 일일 손실 비율, 일수를 입력하여 남은 팔로워 수를 찾음
```

## 로그 변환 (Logarithmic Scale)

어떤 자료들은 차수가 계속 늘어나 직선적으로 관계를 시각화하기 어렵습니다. 이때 로그 변환은 데이터를 압축해 이해하기 쉽도록 만들어줍니다.

## 예제 9. 로그 변환

- 데이터와 밑 (base)를 입력 받아 전체 데이터를 로그 변환하는 알고리즘

```python
import math

def log_scale(data, base):
    return list(map(lambda x: math.log(x, base), data))

log_scale([1, 10, 100, 1000], 10) # 예시로 데이터 리스트와 로그의 밑을 입력하여 로그 스케일로 변환된 값을 찾음
```

```python
import math

def log_scale(data, base):
    return [math.log(x, base) for x in data]

log_scale([1, 10, 100, 1000], 10) # 예시로 데이터 리스트와 로그의 밑을 입력하여 로그 스케일로 변환된 값을 찾음
```

## 평균과 중간값 (Mean & Median)

어떤 데이터를 대표하기 위해서는 평균이나 중간값을 사용합니다. 평균은 전체 데이터의 평균을 의미하며, 중간값은 전체 데이터 중 순서적으로 중간에 위치한 값을 의미합니다. 만약 데이터의 개수가 짝수개라면 중간값은 중간에 위차한 두 값의 평균을 의미합니다. 일반적으로 평균은 이상치에 영향을 크게 받기 때문에 경험적으로 중간값이 어떤 집단을 더 잘 대표하는 편입니다.

## 예제 10. 평균 팔로워 수

- 내가 팔로우 하는 사람들의 팔로워 수의 평균을 계산하는 알고리즘

```python
def average_followers(nums):
    if len(nums) <= 0: return None
    return sum(nums)/len(nums)

average_followers([100, 200, 300]) # 예시로 팔로워 수 리스트를 입력하여 평균 팔로워 수를 찾음
```
